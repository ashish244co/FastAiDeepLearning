```python
import pandas as pd
```

```python
from fastai.vision import *
from pathlib import Path
```

```python
IMAGE_PATH = "../input/plant-pathology-2020-fgvc7/images/"
TEST_PATH = "../input/plant-pathology-2020-fgvc7/test.csv"
TRAIN_PATH = "../input/plant-pathology-2020-fgvc7/train.csv"
SUB_PATH = "../input/plant-pathology-2020-fgvc7/sample_submission.csv"
```

```python
path = Path('/kaggle/input/plant-pathology-2020-fgvc7/')
```

```python
sub = pd.read_csv(SUB_PATH)
test_data = pd.read_csv(TEST_PATH)
train_data = pd.read_csv(TRAIN_PATH)
```

```python
train_data.head()
```

```python
traindf1 = pd.DataFrame()
traindf1['id'] = train_data['image_id']
traindf1['label'] = np.zeros(len(traindf1))
for i in range(0,len(traindf1)):
  if train_data['healthy'][i] == 1:
    traindf1['label'][i] = 'healthy'
  elif train_data['multiple_diseases'][i] == 1:
    traindf1['label'][i] = 'multiple_diseases'
  elif train_data['rust'][i] == 1:
    traindf1['label'][i] = 'rust'
  elif train_data['scab'][i] == 1:
    traindf1['label'][i] = 'scab'
traindf1.head()
```

```python
test_data["image_id"] = test_data["image_id"] + ".jpg"
test_data = ImageList.from_df(test_data, path/'images')
```

```python
src = (ImageList.from_df(traindf1, path, folder='images', suffix='.jpg').split_by_rand_pct(0.2).label_from_df())
```

```python
tfms = get_transforms(flip_vert=True, max_lighting=0.2,max_rotate=180,p_affine=1)
data1 = (src.transform(tfms, size=112))
data1 = data1.add_test(test_data)
data1 = data1.databunch( bs = 64).normalize(imagenet_stats)
```

```python
data1.show_batch()
```

```python
learn = cnn_learner(data1, models.resnet152, metrics=[accuracy , error_rate],model_dir="/tmp/model/")
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, max_lr=slice(1e-2,1e-1))
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, max_lr=slice(1e-6,1e-5))
```

```python
learn.recorder.plot_losses()
```

```python
learn.save('stage1')
```

```python
learn.load('stage1')
```

```python
learn.freeze()
```

```python
data2 = (src.transform(tfms, size=224))
data2 = data2.add_test(test_data)
data2 = data2.databunch( bs = 64).normalize(imagenet_stats)
```

```python
learn.data=data2
```

```python
learn.lr_find()
learn.recorder.plot()
```
